```python
import pandas as pd

# 1. Cargar los datos (ajusta el nombre del archivo si es distinto)
df = pd.read_csv('dataset/churn.csv')
```

```python
# 2. Un vistazo rápido a las primeras 5 filas para entender la estructura
print("--- PRIMERAS 5 FILAS ---")
display(df.head()) 
```

## Detalles de campos

- RowNumber: Numero de filas (no aporta nada)
- CustomerId: Identificador de cliente (parece tener 8 digitos)
- Surname: Apellido del cliente
- CreditScore: Puntaje de credito
- Geography: La ubicacion del cliente
- Gender: El genero del cliente
- Age: La edad del cliente
- Tenure: Numero de años que ha sido cliente del banco
- Balance: Saldo del cliente
- NumOfProducts: Numero de productos que tiene el cliente con el banco
- HasCrCard: Indica si el cliente tiene tarjeta de credito
- isActiveMember: Indica si el cliente es miembro activo del banco
- EstimatedSalary: Sueldo estimado del cliente
- Exited: Indica si el cliente se fue del banco

```python
# 3. Información general (Vital para ver si hay nulos y qué tipo de datos tenemos)
print("\n--- INFORMACIÓN DEL DATASET ---")
df.info()
```

Ningun campo tiene valores nulos.
Pero se observa que los tipos de datos no son adecuados para algunos campos (usa mucha memoria)

```python
# 4. Estadísticas descriptivas (Para ver promedios de edad, sueldos, saldos)
print("\n--- ESTADÍSTICAS DESCRIPTIVAS ---")
display(df.describe())
```

En esta informacion no se debe considerar los campos:
- RowNumber
- CustomerId
- HasCrCard
- IsActiveMember
- Exited

```python
# 5. La métrica de negocio: ¿Cuántos se fueron (1) vs cuántos se quedaron (0)?
print("\n--- TASA DE FUGA ACTUAL (CHURN RATE) ---")
# normalize=True nos da el porcentaje en lugar del conteo
print(df['Exited'].value_counts(normalize=True) * 100)
```

De los 10000 clientes el 79.63% mantiene relacion con el banco y el 20.37% se van del banco.

```python
# 1. Eliminar columnas que no aportan valor analítico (ruido)
# Quitamos Surname por privacidad de datos (buena práctica) y los IDs que no sirven para el modelo
df_clean = df.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
```

```python
# 2. Optimizar memoria y corregir tipos de datos
# Convertir texto a 'category' reduce drásticamente el peso en RAM
df_clean['Geography'] = df_clean['Geography'].astype('category')
df_clean['Gender'] = df_clean['Gender'].astype('category')
```

```python
# Convertir variables binarias para que no se mezclen en la estadística continua
df_clean['HasCrCard'] = df_clean['HasCrCard'].astype('category')
df_clean['IsActiveMember'] = df_clean['IsActiveMember'].astype('category')
# Dejaremos 'Exited' como entero (0 y 1) porque nos servirá para sumar rápidamente la fuga en Power BI y SQL
```

```python
# 3. Comprobar la optimización de memoria
print("--- NUEVA INFORMACIÓN (MEMORIA OPTIMIZADA) ---")
df_clean.info()
```

```python
# 4. Ver estadísticas descriptivas (¡Ahora sí, solo de datos reales!)
print("\n--- ESTADÍSTICAS NUMÉRICAS LIMPIAS ---")
display(df_clean.describe())
```

```python
# 5. Guardar el archivo limpio para la Fase SQL
df_clean.to_csv('Bank_Churn_Cleaned.csv', index=False)
print("\n✅ Archivo 'Bank_Churn_Cleaned.csv' exportado con éxito y listo para la base de datos.")
```
